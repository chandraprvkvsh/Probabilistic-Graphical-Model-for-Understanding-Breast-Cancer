```python
#Part 1

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
df=pd.read_csv('train.csv')
dftest=pd.read_csv('test.csv')

df=df.drop('Unnamed: 0',axis=1)
df=df.drop('Unnamed: 32',axis=1)
dftest=dftest.drop('Unnamed: 0',axis=1)
dftest=dftest.drop('Unnamed: 32',axis=1)
traint=df['diagnosis']
testt=dftest['diagnosis']


df=df.drop('diagnosis',axis=1)
df=df.drop('id',axis=1)
dftest=dftest.drop('diagnosis',axis=1)
dftest=dftest.drop('id',axis=1)


scaler = MinMaxScaler()
trainf= scaler.fit_transform(df)
testf = scaler.fit_transform(dftest)
trainf


logreg=LogisticRegression()
logreg.fit(trainf,traint)
nb=GaussianNB()
nb.fit(trainf,traint)


yprednb=nb.predict(testf)
ypredlr=logreg.predict(testf)
asnb=nb.score(testf,testt)
aslr=logreg.score(testf,testt)
result=[[np.round(asnb,3)],[np.round(aslr,3)]]


import csv
with open('output.csv','w') as out:
    writer=csv.writer(out,delimiter=',')
    writer.writerows(result)
```
